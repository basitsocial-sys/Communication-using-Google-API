# Module 2: Grammar Correction\nDemonstrating how to fix grammatical errors using `language-tool-python`.

```
from transformers import pipeline\n\ntry:\n    corrector = pipeline('text2text-generation', model='vennify/t5-base-grammar-correction')\n    text = 'A sentence with a error in the hitchy.'\n    result = corrector('grammar: ' + text, max_length=128)\n    corrected = result[0]['generated_text']\n    print(f'Original: {text}')\n    print(f'Corrected: {corrected}')\nexcept Exception as e:\n    print(f'Model Error: {e}')\n
```
